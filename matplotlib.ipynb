## Plotting Functions
#### **np.arange** code

```python
# X values.
x = np.arange(0.0,10.0,0.1)
# Starting is included, ending 
# The numbers range from 0.1 to 10,
# incrementing by 0.1 each step.

print (x)
```

#### **np.linspace** code

```python
# In the linspace the third point (step)
# is the number of numbers you want
# between the to numbers (starting and ending) 

# If we change the step to 100
X = np.linspace(0.0,10.0,100)
print (X)
# It prints an unspected output
# that happens This code returns the division between
# the two numbers, and by dividing the range from 0.0 to 10
# into 100 equal parts, the resulting values are 
# numbers with many decimal places.

# If we change the step to 101
Y = np.linspace(0.0,10.0,101)
print (Y)
# This time it divide the range in 101 equal parts, resulting
# in 101 values of 0.1.
# This is an way to write and include the ending number
```

```python
# Y values
y = x ** 2

y
```

```python
# Plot x vs y.
plt.plot(x,y)
```

```python
# Plot x vs y
# Adding labels
# Adding "label='y=x^2'" it adds a label

# instead >>> plt.plot(x,y, label='y=x^2'), can be:
# >>> plt.plot(x,y,label='y=x^2');

# Changing the design
# If add a "." in the next code, it changes
# the design from line to dots.
# intead >>> plt.plot(x,y), can be:
# >>> plt.plot(x,y, ".");

# Changing colour
# Adding "letter dash", it changes the design colour:
# e.g
# plt.plot(x,y, "r-", label='y=x^2'); for red colour.
# combining both codes:
plt.plot(x,y, "g.", label='y=x^2');

# this code add legend to the graphic
plt.legend();
# those adds a label for the elements
plt.xlabel('x')
plt.ylabel('x^2')
# Adding title
plt.title('Adding a title here');
```

#### correct mathematical equations

```python
# Transforming into more correct mathematical equations.
# Adding $ in the code:
plt.plot(x,y, "g.", label='$y=x^2$');

# legend
plt.legend();
# label
plt.xlabel('x')
plt.ylabel('x^2')
# Title
plt.title('See the legend');
```

### Comparing Plots

```python
# Domain of x values.
x = np.linspace (0.0, 100.0, 1001)

# x squared
y1 = x**2

# 2 to the power of x
y2 = 2**x
```

```python
# Plot x vs y1.
plt.plot(x, y1, label='$y = x^2$')

# Plot x vs y2.
plt.plot(x, y2, label='$y = 2^x$')

# Add a legend
plt.legend()

# label
plt.xlabel('x')
plt.ylabel('y')

# Title
plt.title('Two plots at once');
# There is an exponential plot
```

### Integers

###### In *arange*, the code writed in Integer runs integer, but in *linspace* codes always runs in float numbers, unless we add **dtype=int**, **dtype=np.int64**, **dtype=np.int64**.

```python
# Domain of x values.
x = np.arange (1, 101, 1)

# Show
# np.arange prints out int, when imput is int
# and prints float when imput is float.
x
```

```python
# The type
# this code show us the output type.
x.dtype
```

```python
# Domain of x values.
# linspace always print out float types, even when
# imput is int as following
x = np.linspace(1, 100, 100)

# Show
# notice that has dots with numbers
x
```

```python
# This code show us the output types.
x.dtype
```

```python
# To "fix" this just add "dtype=int" into the code.
x = np.linspace(0, 100, 11, dtype=int)

# Show
# Notice there is no dots anymore.
x
```

```python
# The type
# Now the output is integer.
x.dtype
```

```python

# x cubed plus 2 x squared plus 1.

y = x**3 + 2 * x**2 + 1

# Show
y
```

```python
# Type of y.
# "y" will follow the type from the imputs.
# if "x" is an float, or an integer number,
# that will affect "y" type.
y.dtype
```

```python
# Plot x vs y.
plt.plot(x, y,"r.", label='$x^3 + 2x^2 + 1$')
# To plot Integer, is important to 
# plot it with "dots" stead line. 

# Add a legend
plt.legend()

# label
plt.xlabel('x')
plt.ylabel('y')

# Title
plt.title('Ploting Integers');
```
